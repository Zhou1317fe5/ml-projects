#### （2）风险指标量化

```python
import numpy as np
import pandas as pd
data2=pd.read_excel("./data/编码之后的数据.xlsx")
```

A1操作技能
A2制度执行
A3管理能力

A8自然天气

```python
# 添加风险等级列
# 要添加的列名列表
new_columns = ['A1操作技能', 'A2制度执行', 'A3管理能力', 'A4维修保养', 'A5设施设备', 'A6库存管理', 'A7化学反应', 'A8自然天气']

# 添加新列
for col in new_columns:
    data2[col] = ''  # 假设初始值都设置为空
```

```python
data2.info()
```

```python
import pandas as pd
# 风险影响程度列名列表
impact_columns = ['操作技能1', '制度执行1', '管理能力1', '维护保养1', '设施设备1', '库存管理1', '化学反应1', '自然天气1']

# 风险发生概率列名列表
probability_columns = ['操作技能2', '制度执行2', '管理能力2', '维护保养2', '设施设备2', '库存管理2', '化学反应2', '自然天气2']

#风险评估列名列表
risk_columns = ['A1操作技能','A2制度执行','A3管理能力','A4维修保养','A5设施设备','A6库存管理','A7化学反应','A8自然天气']

print()
# 定义风险等级划分规则
# (影响程度,发生概率):等级
risk_matrix = {
    (1, 1): 'L',
    (1, 2): 'L',
    (1, 3): 'L',
    (1, 4): 'M',
    (1, 5): 'M',

    (2, 1): 'L',
    (2, 2): 'L',
    (2, 3): 'M',
    (2, 4): 'M',
    (2, 5): 'M',

    (3, 1): 'L',
    (3, 2): 'M',
    (3, 3): 'M',
    (3, 4): 'M',
    (3, 5): 'H',

    (4, 1): 'M',
    (4, 2): 'M',
    (4, 3): 'M',
    (4, 4): 'H',
    (4, 5): 'H',

    (5, 1): 'M',
    (5, 2): 'M',
    (5, 3): 'H',
    (5, 4): 'H',
    (5, 5): 'H'  

}

# 对每个工作性质进行风险等级评估
for _, row in data2.iterrows():
    for impact_col, probability_col, risk_col in zip(impact_columns, probability_columns,risk_columns):
        impact_level = row[impact_col]
        probability_level = row[probability_col]
        risk_level = risk_matrix.get((impact_level, probability_level), '未定义')
        print(impact_col,probability_col,'---',risk_col,'风险等级:',risk_level)
        data2.at[_,risk_col]=risk_level # 使用.at访问器更新数据框中的单元格值
    print('-------')    
    
```

```python
data3=pd.read_excel('./data/风险等级数据.xlsx')
data3=data3.iloc[:,18:]
data3=data3.T
```

```python
import pandas as pd

# 总数
count = data3.shape[1]

# 统计每行中'L'、'M'、'H'的数量
def count_values(row):
    count = row.value_counts()
    return pd.Series(count)

count_df = data3.apply(count_values, axis=1)

# 计算每行'L'、'M'、'H'的百分比
percentage_df = count_df.div(data3.shape[1]) * 100

# 将count_df和percentage_df合并到原始数据框中
data3['LN'] = count_df['L'] # L个数
data3['L'] = percentage_df['L'] # L百分比
data3['MN'] = count_df['M'] # M个数
data3['M'] = percentage_df['M'] # M百分比
data3['HN'] = count_df['H'] # H个数
data3['H'] = percentage_df['H'] # H百分比
data3['合计']= count # 总数

data3=data3.iloc[:,-7:]
data3
```
